```
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```
def relu(x):
  if x > 0:
    return x
  else:
    return 0
```

```
x = np.linspace(-10, 10, 100)
z=-(x-5)
y = map(relu,x)
u=list(map(relu,x))
v=list(map(relu,z))
plt.plot(x,u,'r',x,v,'b')
```

```
w=[]
for i in range(0,len(x)):
  w+=[u[i]+v[i]]
plt.plot(x,w)
```
